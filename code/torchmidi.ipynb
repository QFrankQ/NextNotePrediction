### Torch MIDI Examples
This notebook contains examples showing how torch tensors and torch data files distributed with the assignment can be converted to MIDI files and played. To run this notebook, you will first need to install the dependencies using pip install -r requirements.txt

```python
%load_ext autoreload
%autoreload 2
```

```python
import torch
import miditools
import glob
```

### Define a score using a torch tensor, save and play

```python
X = torch.zeros(8,4).type(torch.int)

X[:,0] = 300 #Set start times
X[:,1] = 200 #Set durations
X[7,1] = 500
X[:,2] = torch.tensor([64, 66, 68, 69, 71, 73, 75, 76]) #Set durations
X[:,3] = 50 #Set volume

miditools.play_torch(X)
```

### Load a random song from the training set and display player

```python
files=glob.glob("../data/songs/train/*/*/*/*.pt")
num_files = len(files)
print(f"Found {num_files } torch data files.")
i = torch.randint(0,num_files-1,(1,)).item()
X = torch.load(files[i])
print(type(X[0,0].item()))
print(X[:10])
miditools.play_torch(X)
```

### Load a random song context from the test set and display player

```python
X = torch.load("../data/songs/test.pt")
Ntest = X.shape[0]
i = torch.randint(0,Ntest,(1,)).item()
print(f"Found {Ntest} test cases.")
print(X.size())
miditools.play_torch(X[i,:,:])
```
